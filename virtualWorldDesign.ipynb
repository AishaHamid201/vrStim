```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
npts = [50,100,200]

fig, ax = plt.subplots(1,len(npts),figsize=(12,4))

for i, npt in enumerate(npts):
    
    locs = (np.random.uniform(0, 1, npt),np.random.uniform(0, 1, npt))

    ax[i].plot(locs[0],locs[1], 'k.')
    ax[i].set_aspect('equal')
    ax[i].set_title('{} points'.format(npt))
ax[0].set_ylabel('Uniform distribution\n $\min={}, \max={}$'.format(0,1) )

fig.savefig('placementMap_uniform.pdf')
```

```python
npts = [50,100,200]
mu = 0.4
sigma = 0.2

fig, ax = plt.subplots(1,len(npts),figsize=(12,4))

for i, npt in enumerate(npts):
    
    locs = (np.random.normal(mu, sigma, npt),np.random.normal(mu, sigma, npt))

    ax[i].plot(locs[0],locs[1], 'k.')
    ax[i].set_aspect('equal')
    ax[i].set_title('{} points'.format(npt))
    ax[i].set_xlim(0,1)
    ax[i].set_ylim(0,1)
    
ax[0].set_ylabel('Normal distribution\n $\mu={}, \sigma={}$'.format(mu, sigma) );
fig.savefig('placementMap_normal.pdf')
```
